```python
import numpy
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
iris = (sklearn.datasets.load_iris())
print(iris)
```

```python
print(type(iris["data"]))
print(iris["data"].shape)

print(type(iris["target"]))
print(iris["target"].shape)

features = iris["data"]
labels = iris["target"]

feature_names = iris["feature_names"]
label_names = iris["target_names"]

print(features)
print(features.shape)
print(labels)
print(labels.shape)
print(feature_names)
print(label_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.5)
print(X_train)
print(X_train.shape)
print(y_train)
print(y_train.shape)
# print(len([y for y in y_train if y == 0]))
# print(len([y for y in y_train if y == 1]))
# print(len([y for y in y_train if y == 2]))
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
print(clf)

y_pred = clf.predict(X_test)
print(y_test)
print(y_pred)
```

```python
score = accuracy_score(y_test, y_pred)
print(score * 100)
```

```python
plt.figure(figsize=(16, 12))
plot_tree(clf, filled=True, feature_names=feature_names, class_names=label_names,fontsize=10)
plt.show()
```
